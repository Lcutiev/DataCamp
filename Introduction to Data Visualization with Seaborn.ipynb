## **Create and interpret a box plot**

Letâ€™s take the first box plot i.e, blue box plot of the figure and understand these statistical things:
+ Bottom black horizontal line of blue box plot is minimum value
+ First black horizontal line of rectangle shape of blue box plot is First quartile or 25%
+ Second black horizontal line of rectangle shape of blue box plot is Second quartile or 50% or median.
+ Third black horizontal line of rectangle shape of blue box plot is third quartile or 75%
+ Top black horizontal line of rectangle shape of blue box plot is maximum value.
+ Small diamond shape of blue box plot is outlier data or erroneous data.

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]
```

```python
# Create a box plot and set the order of the categories
```

```python
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="box",
            order=study_time_order)
```

```python
sns.boxplot(x="study_time", y="G3",
            data=student_data,
            order=study_time_order)
```

### Changing the whiskers using `whis`

+ By default,the whiskers extend to 1.5* the interquartile range 
+ Make them extend to 2.0*IQR: `whis=2.0`
+ Show the 5th and 95th percentiles: `whis=[5, 95]`
+ Show min and max values: `whis=[0, 100]`


```python
# Set the whiskers to 0.5 * IQR
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            order=['yes','no'],
            sym='',
            whis=0.5)
```

```python
# Extend the whiskers to the 5th and 95th percentile
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            order=['yes','no'],
            sym='',
            whis=[5,95])
```

```python
sns.boxplot(x="romantic", y="G3",
            data=student_data,
            order=['yes','no'],
            sym='',
            whis=[5,95])
```

```python
# Set the whiskers at the min and max values
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            order=['yes','no'],
            sym='',
            whis=[0,100])
```

**Comment:**
+ The median grade is the same between these two groups, but the max grade is higher among students who are not in a romantic relationship.

+ Seaborn plot functions create two different types of objects: `FacetGrid` objects and `AxesSubplot` objects. 
+ The method for adding a title to your plot will differ depending on the type of object it is.

## Adding a title to a FacetGrid object

```python
# Create scatter plot
weight_power = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")
# Add a title "Car Weight vs. Horsepower"
weight_power.fig.suptitle('Car Weight vs. Horsepower')
```

```python
# Create scatter plot
weight_power = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")
# Adjusting height of title in FacetGrid
weight_power.fig.suptitle('Car Weight vs. Horsepower',
                         y=1.03)
```

## Rotating x-tick labels

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)
# Rotate x-tick labels
plt.xticks(rotation=90)
```
