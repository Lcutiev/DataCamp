## 13. nh gi cht lng hot ng giao dch ca cc ngn hng trong cc thng

```python
# Gi s trong mt thng s giao dch (c gi i v nhn v) ca mt ngn hng 
# nh hn 3000 th c coi l 'km', t 30000 n 40000 c coi l 'bnh thng', cn trn 4000 c coi l 'tt'
import numpy as np
trans_month = df.groupby(['MONTH','FROM_CODE', 'TO_CODE'])['TRANS'].sum()
conditions = [(trans_month < 3000),
              (trans_month >=3000) & (trans_month <= 4000),
              (trans_month > 4000)
]

choices = ['Bad', 'Normal', 'Good']

df['RATE'] = np.select(conditions, choices)
df[['FROM','MONTH','RATE']].head()
```

## 14. Tm s giao dch trong thng theo 3 khung thi gian: sng(0-12), chiu(12-18), ti(18-24)

```python
# Convert time from 12 hour to 24 hour format
df['PERIOD'] = pd.to_datetime(df['HOURS']).dt.strftime('%H:%M:%S')
df.head()
```

```python
morning = df[(df['PERIOD'] >= '00:00:00') & (df['PERIOD'] <= '12:00:00')]
morning[['MONTH', 'FROM', 'PERIOD', 'TRANS']]
```

```python
evening = df[(df['PERIOD'] >= '12:00:00') & (df['PERIOD'] <= '18:00:00')]
evening[['MONTH', 'FROM', 'PERIOD', 'TRANS']]
```

```python
night = df[(df['PERIOD'] >= '18:00:00') & (df['PERIOD'] <= '24:00:00')]
night[['MONTH', 'FROM', 'PERIOD', 'TRANS']]
```
