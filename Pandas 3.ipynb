### Hợp nhất các dữ liệu DataFrame

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],'hire_date': [2004, 2008, 2012, 2014]})
print(df1); 
print(df2)
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
pd.merge(df1, df2, on='employee')
```

```python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],'salary': [70000, 80000, 120000, 90000]})
df3
```

```python
pd.merge(df1, df3, left_on='employee', right_on='name')
```

### Pivot Table

```python
data.head()
```

```python
# Xem điểm trung bình Toán theo Quê quán, Giới tính
data.groupby(['Quê quán', 'Giới tính'])['Toán'].aggregate('mean')
```

```python
# .unstack : đổi cột và hàng cho nhau cho dễ nhìn dữ liệu
data.groupby(['Quê quán', 'Giới tính'])['Toán'].aggregate('mean').unstack() # pivot làm bằng tay sử dụng groupby
```

```python
data.pivot_table('Toán', index='Quê quán', columns='Giới tính')
```

```python
data.pivot_table('Toán', index='Giới tính', columns='Quê quán')
```
