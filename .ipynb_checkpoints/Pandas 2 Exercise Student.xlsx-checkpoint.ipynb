### Tính số lượng sinh viên có điểm thi từng môn trong từng phổ điểm

```python
# Chia phổ điểm từ 0-10 thành 10 đoạn (0-1, 1-2, 2-3,..9-10). Tính số lượng sinh viên có điểm thi từng môn trong từng phổ điểm trên
```

```python
# Tính điểm trung bình các môn của từng học sinh và cho cột điểm này vào bộ dữ liệu. 
# Biết rằng các môn Toán, Văn, và tiếng Anh hệ số 2.0, các môn còn lại hệ số 1.0
```

```python
subjects = ['Toán', 'Lý', 'Hoá', 'Sinh', 'Văn', 'Địa', 'Tiếng Anh']
dtb = df[subjects]
dtb
```

```python
tva = ['Toán', 'Văn', 'Tiếng Anh']
lhs = ['Lý', 'Hoá', 'Sinh', 'Địa']
dtb = (df[tva].sum(axis=1) * 2 + df[lhs].sum(axis=1)) / 10
dtb
```

```python
df['Trung bình'] = dtb
```

```python
df
```

### Tính điểm cho các khối A, A1, B, C, D

```python
A = df[['Toán', 'Lý', 'Hoá']].sum(axis=1)
df['Khối A'] = A
```

```python
A1 = df[['Toán', 'Lý', 'Tiếng Anh']].sum(axis=1)
df['Khối A1'] = A1
```

```python
B = df[['Toán', 'Hoá', 'Sinh']].sum(axis=1)
df['Khối B'] = B
```

```python
C = df[['Văn', 'Sử', 'Địa']].sum(axis=1)
df['Khối C'] = C
```

```python
tv = ['Toán', 'Văn']
#D = df[['Toán', 'Văn', 'Anh']].sum(axis=1)
D = (df['Tiếng Anh'] * 2 + df[tv].sum(axis=1)) / 10
df['Khối D'] = D
```

```python
df
```

### Xét tuyển đầu vào các khối

```python
# fail or pass A: 23.5
passA = []
for i in df['Khối A']:
    if i >= 23.5:
        passA += ['pass']
    else:
        passA += ['fail']

df['passA'] = passA
```

```python
# fail or pass A1: 22
passA1 = []
for i in df['Khối A1']:
    if i >= 22:
        passA1 += ['pass']
    else:
        passA1 += ['fail']

df['passA1'] = passA1
```

```python
# fail or pass B: 21.5
passB = []
for i in df['Khối B']:
    if i >= 21.5:
        passB += ['pass']
    else:
        passB += ['fail']

df['passB'] = passB
```

```python
# fail or pass C: 22.5
passC = []
for i in df['Khối C']:
    if i >= 22.5:
        passC += ['pass']
    else:
        passC += ['fail']

df['passC'] = passC
```

```python
# fail or pass D: 27.5
passD = []
for i in df['Khối D']:
    if i >= 27.5:
        passD += ['pass']
    else:
        passD += ['fail']

df['passD'] = passD
```

### Xếp loại học lực học sinh

```python
import numpy as np

def judge(x):
    if any(x > 8.0):
        return 'Xuất sắc'
    else:
        return 'Trung bình'

tb = np.array(df['Trung bình'])
if tb.any() > 9.0:
    df['Học lực'] = df[['Toán', 'Lý', 'Hoá']].apply(judge, axis=1)
```
