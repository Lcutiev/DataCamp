## Dùng chỉ số cột và dòng để lọc dữ liệu

```python
import pandas as pd
students = pd.read_csv('students.csv')
students.head()
```

```python
# 1: lọc các cột dữ liệu
students[students.columns[5:]]
```

```python
# 2: lọc các cột dữ liệu > fancy index
students[['Toán', 'Lý', 'Hoá', 'Sinh', 'Văn', 'Sử', 'Địa', 'Tiếng Anh']]
```

```python
# Lọc dữ liệu theo dòng: loc & iloc
students.loc[1:5]
```

```python
# Lọc dữ liệu theo cột và dòng
students[students.columns[5:]][:4]
```

```python
# iloc: lấy 3 - dòng đầu tiên, 2 -cột đầu tiên
# Lấy một khoảng data nhỏ của 1 data lớn

students.iloc[:3, :2]
```

```python
students.iloc[:4, 5:]
```

## Lọc bằng phương thức filter

```python
# Lọc các cột
students.filter(['Toán', 'Lý', 'Hoá']) # Liệt kê tường minh
```

```python
students.filter(students.columns[5:])
```

```python
# Lọc các dòng rời rạc
# Lọc theo dòng nhớ bỏ axis = 0
students.filter([0,1,3,5,7],axis=0)
```

```python
students.index[students.index%2!=0]
```

```python
# Lọc theo dòng có điều kiện
students.filter(students.index[students.index%2!=0], axis = 0)
```

```python
# Lọc theo cột và dòng
students.filter(['Toán', 'Lý', 'Hóa'])[:5]
```

```python
students.filter(['Toán', 'Lý', 'Hóa']).iloc[:5]
```

```python
students.filter(['Toán', 'Lý', 'Hóa']).loc[:5]
```

## Lọc có điều kiện

```python
# Lọc theo điều kiện đơn
# Các toán tử có thể dùng: ==, >, >=, <, <=, !=..
students[students['Toán']>7]
```

```python
# Lọc theo nhiều điều kiện
# & - và, | - hoặc, ~ phủ định
students[(students['Toán']>7) & (students['Văn']>=6.5)]
```

```python
students[~((students['Toán']>7) & (students['Văn']>=6.5))]
```

```python
# Lọc có điều kiện bằng phương thức query
# Cho kết quả, performance tốt hơn
students.query("Toán > 7 and Văn >=6.5")
```

```python
students.query("Toán > 7 and not Văn >=6.5")
```

```python
# Nhiều lựa chọn sử dụng np.select
import numpy as np
xs = students['Toán'] >=9
gioi = (students['Toán'] >=8) & (students['Toán']<9)
gioi = (students['Toán'] >=8) & (students['Toán']<9)
kha = (students['Toán'] >=6) & (students['Toán']<8)
tb = (students['Toán'] >=5) & (students['Toán']<7)
conditions = [xs, gioi, kha, tb]
xl = ['Xuất sắc', 'Giỏi', 'Khá', 'Trung bình']
students['XL Toán'] = np.select(conditions, xl, default='Kém')
students
```

### Tìm số lượng và thông tin những sinh viên không đỗ bất cứ khối nào.

```python
# Tìm số lượng và thông tin những sinh viên không đỗ bất cứ khối nào.
# Gỉa sử điểm xét tuyển đầu vào của một trường đại học với tiêu chí khối A: 23.5, A1: 22, B: 21.5, C:22.5, D: 27.5
```

```python
students['Khối A']= students[['Toán', 'Lý', 'Hóa']].sum(axis=1)
students['Khối A1'] = students[['Toán', 'Lý', 'Tiếng Anh']].sum(axis=1)
students['Khối B']= students[['Toán', 'Hóa', 'Sinh']].sum(axis=1)
students['Khối C'] = students[['Văn', 'Sử', 'Địa']].sum(axis=1)
students['Khối D'] = students['Tiếng Anh'] * 2 + students[['Toán', 'Văn']].sum(axis=1)
```

```python
students.rename(columns={'Khối A': 'A', 'Khối A1': 'A1', 'Khối B': 'B', 'Khối C': 'C', 'Khối D': 'D'},inplace=True)
```

```python
students
```

```python
conditions = (students.A >=23.5) | (students.A1 >=22) | (students.B >= 21.5) | (students.C >= 22.5) | (students.D >= 27.5)
```

```python
import numpy as np
conditions = (students.A >=23.5) | (students.A1 >=22) | (students.B >= 21.5) | (students.C >= 22.5) | (students.D >= 27.5)

students['Xét tuyển'] = np.where(conditions, 'Đậu', 'Rớt')
students
```

```python
# Tìm số lượng và thông tin những sinh viên không đỗ bất cứ khối nào
students[students['Xét tuyển'] =='Rớt'].value_counts()
```

### Hãy cho biết những học sinh nào đỗ vào trường H và có được chọn vào lớp tài năng không?

```python
# Một trường đại học H tuyển sinh khối A với mức điểm là 18.5
# Trường cũng tuyển sinh 1 lớp tài năng với số điểm đầu vào là 23 và số điểm toán từ điểm 8
đkH = (students['Khối A']>= 18.5) & (students['Toán'] >=8) & (students['Khối A'] >= 23)
students['Lớp tài năng'] = np.where(đkH, 'Chọn', 'Tạch')
students[students['Lớp tài năng'] =='Chọn']
```
