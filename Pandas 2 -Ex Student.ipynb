```python
# Chia phổ điểm từ 0-10 thành 10 đoạn
# Tính số lượng sinh viên có điểm thi từng môn trong từng phổ điểm trên
spec = {(i, i +1): 0 for i in range(11)}
for i in df['Toán'].values:
    if i <1:
        spec[(0, 1)]+=1
    elif i<2:
        spec[(1, 2)]+=1
    elif i <3:
        spec[(2, 3)]+=1
    elif i <4:
        spec[(3, 4)]+=1
    elif i <5:
        spec[(4, 5)]+=1
    elif i <6:
        spec[(5, 6)]+=1
    elif i <7:
        spec[(6, 7)]+=1
    elif i <8:
        spec[(7, 8)]+=1
    elif i <9:
        spec[(8, 9)]+=1
    else:
        spec[(9, 10)]+=1
spec  
```

### Xét tuyển đầu vào các khối
